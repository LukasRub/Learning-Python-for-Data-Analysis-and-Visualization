```python
def male_female_child(passenger):
    age, sex = passenger
    if age < 16:
        return 'child'
    else:
        return sex
```

```python
titanic_df['person'] = titanic_df[['Age','Sex']].apply(male_female_child, axis=1)
```

```python
deck = titanic_df['Cabin'].dropna()
```

```python
deck.head()
```

```python
levels = []

for level in deck:
    levels.append(level[0])
    
cabin_df = DataFrame(levels)
cabin_df.columns = ['Cabin']
sns.factorplot('Cabin', data=cabin_df, palette='winter_d', kind='count')
```

```python
cabin_df = cabin_df[cabin_df.Cabin != 'T']
```

```python
order = np.sort(cabin_df.Cabin.unique())
```

```python
sns.factorplot('Cabin', data=cabin_df, kind='count', order=order, palette='summer')
```

```python
def find_deck(cabin):
    if not pd.isnull(cabin):
        cabin = cabin[0]
    return cabin

titanic_df['Deck'] = titanic_df['Cabin'].apply(find_deck)
```

```python
deck_df = titanic_df[titanic_df['Deck'].notnull()]
deck_df = titanic_df[titanic_df['Deck'] != 'T']
deck_df.head()
```

```python
sns.factorplot('Deck', 'Survived', hue='Sex', data=deck_df, palette = 'winter', size=8, aspect=2, order=order, kind='violin')
```

```python
sns.factorplot('Alone', 'Survived', hue='Sex', data=deck_df, palette = 'winter', size=8, aspect=2, kind='point')
```
