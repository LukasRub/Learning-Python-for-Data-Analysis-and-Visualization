```python
# The usual imports
import numpy as np
import pandas as pd
from numpy.random import randn

# Import the stats library from numpy
from scipy import stats

# Import plotting modules
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data1 = randn(100)
data2 = randn(100)
```

```python
sns.boxplot(data=[data1, data2])
```

```python
sns.boxplot(data=[data1, data2], whis=np.inf)
```

```python
sns.boxplot(data=data1, orient='horizontal')
```

```python
# Normal Dist
data1 = stats.norm(0,5).rvs(100)

# Two gamma dist. concatenated together
data2 = np.concatenate([stats.gamma(5).rvs(50)-1, stats.gamma(5).rvs(50) * -1])

sns.boxplot(data=[data1, data2], whis=np.inf)
```

```python
sns.violinplot(data=[data1, data2], inner='quartile')
```

```python
sns.violinplot(data=data2, bw=0.1)
```

```python
sns.violinplot(data=data1, inner='stick')
```
