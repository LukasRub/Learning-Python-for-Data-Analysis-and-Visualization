```python
candidates = donor_df.cand_nm.unique()
candidates
```

```python
candidates_affil = {}
for c in candidates:
    if "Obama" in c:
        candidates_affil[c] = "Democrat"
    else:
        candidates_affil[c] = "Republican"
candidates_affil
```

```python
donor_df = donor_df[donor_df.contb_receipt_amt > 0]
donor_df['Party'] = donor_df.cand_nm.map(candidates_affil)

donor_df.head()
```

```python
donor_df.groupby('cand_nm')['contb_receipt_amt'].count()
```

```python
donor_df.groupby('cand_nm').contb_receipt_amt.sum()
```

```python
candidate_amount = donor_df.groupby('cand_nm').contb_receipt_amt.sum()

i = 0

for donation in candidate_amount:
    print('The candidate %s raised $%.2f' %(candidate_amount.index[i], donation))
    i += 1
```

```python
candidate_amount.plot(kind='bar')
```

```python
donor_df.groupby('Party').contb_receipt_amt.sum().plot(kind='bar')
```

```python
occupation_df = donor_df.pivot_table('contb_receipt_amt',
                                    index='contbr_occupation',
                                    columns='Party',
                                    aggfunc='sum')
```

```python
occupation_df.head()
```

```python
occupation_df.info()
```

```python
occupation_df = occupation_df[occupation_df.sum(1) > 1000000]
```

```python
occupation_df.plot(kind='barh', figsize=(20,20), layout='vertical', colormap='seismic')
```

```python
occupation_df.drop(['INFORMATION REQUESTED PER BEST EFFORTS', 'INFORMATION REQUESTED'], axis=0, inplace=True)
```

```python
occupation_df.loc['CEO'] = occupation_df.loc['CEO'] + occupation_df.loc['C.E.O.']
occupation_df.drop('C.E.O.', axis=0, inplace=True)
```

```python
occupation_df.plot(kind='barh', figsize=(20,20), layout='vertical', colormap='seismic')
```
