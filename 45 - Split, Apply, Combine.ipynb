```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
dframe = pd.read_csv('winequality-red.csv', sep=';')
dframe.head()
```

```python
def ranker(df):
    df['alc_content_rank'] = np.arange(len(df)) + 1
    return df
```

```python
dframe.sort_values(by='alcohol', ascending=True, inplace=True)
```

```python
dframe = dframe.groupby('quality').apply(ranker)
```

```python
dframe.head()
```

```python
num_of_qual = dframe['quality'].value_counts()
num_of_qual
```
